```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
path = "data"
os.listdir(path)
```

```python
df = pd.read_csv("data\서울시버스정류소위치정보(20240507).csv", encoding="CP949")
df
```

```python
# csv파일의 좌표를 기준으로 gpd변환
geometry = gpd.points_from_xy(df["X좌표"], df["Y좌표"])
gdf = gpd.GeoDataFrame(df, geometry= geometry, crs=4326)
gdf.head()
```

```python
gdf.to_file("data\서울_버스정류소.gpkg", driver="GPKG", encoding="utf-8")
```
